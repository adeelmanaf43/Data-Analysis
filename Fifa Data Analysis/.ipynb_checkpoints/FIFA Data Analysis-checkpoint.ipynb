## <center> FIFA WORLD CUP DATA ANALYSIS </center>

```python
#importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading world cup data from csv file
data = pd.read_csv("Data/WorldCups.csv")
data.head()
```

```python
#reading world cup matches data from csv file
matches = pd.read_csv("Data/WorldCupMatches.csv")
matches.head()
```

```python
#reading world cup players data from csv file
players = pd.read_csv("Data/WorldCupPlayers.csv")
players.head()
```

#### Final Match Results

```python
#merging matches and players data
full = pd.merge(matches, players, how = 'outer', on = ['RoundID', "MatchID"])
finals = full.query("Stage == 'Final'")
finals = finals.drop_duplicates()
```

```python
#getting winner of each world cup
finals_res = finals.drop_duplicates(subset = "Year").reset_index(drop = True)
def winner():
    win = []
    for i in range(finals_res.shape[0]):
        if finals_res["Home Team Goals"][i] > finals_res["Away Team Goals"][i]:
            win.append(finals_res["Home Team Name"][i])
        elif finals_res["Home Team Goals"][i] < finals_res["Away Team Goals"][i]:
            win.append(finals_res["Away Team Name"][i])
        else:
            win.append(finals_res["Win conditions"][i].split(" ")[0])
    return win
winner = winner()
```

```python
#printing result of Finals
finals_res["winner"] = winner
finals_results = finals_res[["Year", "Datetime", "Stadium", "City", "Home Team Name", "Home Team Goals", "Away Team Goals", "Away Team Name", "winner"]]
finals_results
```

#### Top Goal Scorer in Each World Cup

```python
#getting Top Goal Scorer in Each World Cup
import warnings
warnings.filterwarnings("ignore")
full2 = full.drop_duplicates()
goals = full2[(full2["Event"].str.contains("G", na = False)) | (full2["Event"].str.contains("P", na = False))]
#goals = full2[full2["Event"].str.contains("G", na = False)]
goals["Goals"] =goals["Event"].apply(lambda x: x.count("G") + x.count("P"))
```

```python
goals.groupby(["Year", "Player Name"], as_index = False)["Goals"].sum().sort_values(by = "Goals", ascending = False).drop_duplicates(subset = "Year", keep = "first").sort_values(by = "Year").reset_index(drop = True)
```

#### Most Goals in Finals and Semi Finals

```python
finals_data = goals[goals["Stage"].isin(["Final", "Semi Final"])]
finals_data.groupby("Player Name", as_index = False)["Goals"].sum().sort_values(by = "Goals", ascending = False)
```

#### All time top Goal Scorer

```python
#getting all time top goal scorer
goals.groupby(["Player Name", "Team Initials"], as_index = False)["Goals"].sum().sort_values(by = "Goals", ascending = False)[:10][["Player Name", "Goals"]].reset_index(drop = True)
```

#### Top Goal Scorer for each Country

```python
#top goal scorer for each country
goals_team = goals.groupby(["Team Initials", "Player Name"], as_index = False)["Goals"].sum().sort_values(by = "Goals", ascending = False).drop_duplicates(subset = "Team Initials", keep = "first").reset_index(drop = True)
```

#### Team Records in all World Cups

```python
home = matches["Home Team Name"].value_counts().sort_index()
away = matches["Away Team Name"].value_counts().sort_index()
m = home + away
m = pd.DataFrame(m).reset_index().rename(columns = {"index":"Country", 0:"Played"})
m["Played"] = m["Played"].astype("Int64")
```

```python
matches = matches.dropna()
matches.shape
```

```python
matches = matches.reset_index(drop = True)
def wins():
    wins = []
    for i in range(matches.shape[0]):
        if matches["Home Team Goals"][i] > matches["Away Team Goals"][i]:
            wins.append(matches["Home Team Name"][i])
        elif matches["Home Team Goals"][i] < matches["Away Team Goals"][i]:
            wins.append(matches["Away Team Name"][i])
        elif matches["Home Team Goals"][i] == matches["Away Team Goals"][i]:
            if matches["Win conditions"][i].split(" ")[0] != "":
                wins.append(matches["Win conditions"][i].split(" ")[0])
        else:
            wins.append(matches["Home Team Name"][i] + matches["Away Team Name"][i])
    return wins
winss = wins()
```

```python
dic = {}
for j in winss:
    if j not in dic:
        dic[j] = 1
    else:
        dic[j] += 1
```

```python
sorted_dict = {key: value for key, value in sorted(dic.items())}
country = sorted_dict.keys()
wins = sorted_dict.values()
dd = pd.DataFrame()
dd["Country"] = country
dd["Wins"]  = wins
match_ = pd.merge(m, dd, how = "left", on = "Country")
```

```python
def draw():
    dic2 = {}
    for i in range(matches.shape[0]):
        if matches["Home Team Goals"][i] == matches["Away Team Goals"][i] and matches["Win conditions"][i] == " ":
            if matches["Home Team Name"][i] not in dic2:
                dic2[matches["Home Team Name"][i]] = 1
            else:
                dic2[matches["Home Team Name"][i]] += 1
            if matches["Away Team Name"][i] not in dic2:
                dic2[matches["Away Team Name"][i]] = 1
            else:
                dic2[matches["Away Team Name"][i]] += 1
    return dic2
draws = draw()
```

```python
sorted_draws = {key: value for key, value in sorted(draws.items())}
country = sorted_draws.keys()
dra = sorted_draws.values()
dr = pd.DataFrame()
dr["Country"] = country
dr["Draws"]  = dra
match_ = pd.merge(match_, dr, how = "left", on = "Country")
```

```python
def Loss():
    loss = []
    for i in range(matches.shape[0]):
        if matches["Home Team Goals"][i] < matches["Away Team Goals"][i]:
            loss.append(matches["Home Team Name"][i])
        elif matches["Away Team Goals"][i] < matches["Home Team Goals"][i]:
            loss.append(matches["Away Team Name"][i])
        elif matches["Home Team Goals"][i] == matches["Away Team Goals"][i] and matches["Win conditions"][i] != " ":
            if matches["Win conditions"][i].split(" ")[0] == matches["Home Team Name"][i]:
                loss.append(matches["Away Team Name"][i])
            elif matches["Win conditions"][i].split(" ")[0] == matches["Away Team Name"][i]:
                loss.append(matches["Home Team Name"][i])
    return loss
            
loss = Loss()
```

```python
losses = {}
for j in loss:
    if j not in losses:
        losses[j] = 1
    else:
        losses[j] +=1
sorted_losses = {key: value for key, value in sorted(losses.items())}
country = sorted_losses.keys()
losss = sorted_losses.values()
lo = pd.DataFrame()
lo["Country"] = country
lo["Losses"]  = losss
match_ = pd.merge(match_, lo, how = "left", on = "Country")
```
